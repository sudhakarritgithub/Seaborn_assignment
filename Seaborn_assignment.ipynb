```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True)
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'age' and 'fare' by 'pclass'
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram for 'price' with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species'
sns.pairplot(data=iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)
plt.title("Heatmap of Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
